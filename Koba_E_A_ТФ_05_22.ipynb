```python
import numpy as np
import matplotlib.pyplot as plt
import CoolProp.CoolProp as ct
p=1.0133*10**5 #Давление 1 атм
T=273.15 #Температура
V=64 #Обьем
R=8.3145 #Универсальная газовая постоянная 
M=0.016 #Молярная масса метана 
T=np.linspace(50,150,1000) #Параметр температуры 
m1=(p*V*M)/(R*T) #Расчет массы 
D=ct.PropsSI("D","T",T,"P",p,"CH4") #Плотность
m2=D*V #Масса через CoolProp
plt.plot(T,m2,"g")
plt.plot(T,m1,"b")
plt.scatter(111.65,ct.PropsSI("D","T",111.65,"P",p,"CH4"))
plt.title("Зависимость m от T при постоянном давлении")
plt.xlabel('T')
plt.ylabel("m")
plt.grid()
plt.show()


```
